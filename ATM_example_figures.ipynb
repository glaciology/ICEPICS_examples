Viewing Operation Icbridge ATM data for Renland and Sukkertopen- Examples for ICEPICs proposal

```python
# Modules
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_directory_renland = '/Users/bo/dartmouth_local/data/ATM/renland/'
data_directory_sukkertopen = '/Users/bo/dartmouth_local/data/ATM/sukkertopen/'
```

These data files come directly from NSIDC, and were downloaded/discovered though the OIB data portal:
https://nsidc.org/icebridge/portal/map

```python
# Read in Renland data
my_data14 = np.genfromtxt(data_directory_renland + 'ILATM2_20140421_155653_smooth_nadir3seg_50pt.csv', delimiter=',')
my_data18 = np.genfromtxt(data_directory_renland + 'ILATM2_20180421_154900_smooth_nadir3seg_50pt.csv', delimiter=',')

# Here only trying to slice out the Nadir points for visualization.  
indices14 = np.where(my_data14[:,10] == 0);
indices18 = np.where(my_data18[:,10] == 0);

nadir14 = my_data14[indices14,:]
nadir18 = my_data18[indices18,:]
```

```python
plt.rcParams.update({'font.size': 14})
fig = plt.figure(figsize=(10,6))

line14 = plt.plot(nadir14[0,:,2],nadir14[0,:,3], label='21 April 2014')
line18 = plt.plot(nadir18[0,:,2],nadir18[0,:,3], label='21 April 2018')

plt.xlim(333.3, 333.35)
plt.ylim(2365, 2395)
plt.setp(line14,linewidth=3)
plt.setp(line18,linewidth=3)

plt.xlabel('Longitude ($^o$)', fontsize=14, fontweight='bold')
plt.ylabel('Elevation (m)', fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```

```python
# Read in Sukkertopen data
my_data10c = np.genfromtxt(data_directory_sukkertopen + 'ILATM2_20100515_154024_smooth_nadir3seg_50pt.csv', delimiter=',');
my_data10d = np.genfromtxt(data_directory_sukkertopen + 'ILATM2_20100515_154419_smooth_nadir3seg_50pt.csv', delimiter=',');
my_data14 = np.genfromtxt(data_directory_sukkertopen + 'ILATM2_20140416_180703_smooth_nadir3seg_50pt.csv', delimiter=',');
```

```python
# the slice I wanted in 2010 comes from two different files- combine them.
my_data10 = np.concatenate((my_data10c, my_data10d))

# Here only trying to slice out the Nadir points for visualization.  
indices10 = np.where(my_data10[:,10] == 0);
indices14b = np.where(my_data14[:,10] == 0);

nadir10 = my_data10[indices10,:]
nadir14b = my_data14[indices14b,:]
```

```python
fig = plt.figure(figsize=(10,6))
line10 = plt.plot(nadir10[0,:,2],nadir10[0,:,3], label='15 May 2010')
line14b = plt.plot(nadir14b[0,:,2],nadir14b[0,:,3], label='16 April 2014')
plt.xlim(308, 308.1)
plt.ylim(1620, 1650)

plt.setp(line14b,linewidth=3)
plt.setp(line10,linewidth=3)

plt.xlabel('Longitude ($^o$)', fontsize=14, fontweight='bold')
plt.ylabel('Elevation (m)', fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```
